### STEP THREE: Write your Flux query 

```python
query='''
from(bucket: "System")
  |> range(start:-30d, stop: now())
  |> filter(fn: (r) => r._measurement == "cpu")
  |> filter(fn: (r) => r._field == "usage_user")
  |> filter(fn: (r) => r.cpu == "cpu-total")
  |> limit (n:10)
'''
```

### STEP FOUR: Query InfluxDB and return a DataFrame

```python
client = InfluxDBClient(url=urlCloud, token=my_token, org=my_org, debug=False)

system_stats_df = client.query_api().query_data_frame(org=my_org, query=query)
system_stats_df.head()
```

### Using Flux to prepare a data out for conversion to a Pandas Dataframe 

If we want to query more than one field, we will have multiple values in the "_field" column. The shape of the resulting DataFrame isn't optimal for applying manipulating data with Pandas.  Instead we want to perform a pivot with Flux so that we can seperrate out the fields into seperate columns. Dropping additional columns can be useful to produce a clean dataframe as well. Now we can easily plot our values. 


```python
query_with_multiple_fields ='''
from(bucket: "System")
  |> range(start:-30d, stop: now())
  |> filter(fn: (r) => r._measurement == "cpu")
  |> filter(fn: (r) => r._field == "usage_user")
  |> filter(fn: (r) => r.cpu == "cpu-total" or r.cpu == "cpu0" or r.cpu == "cpu1")
  |> limit (n:10)
'''
```

```python
client = InfluxDBClient(url=urlCloud, token=my_token, org=my_org, debug=False)

new_system_stats_df = client.query_api().query_data_frame(org=my_org, query=query_with_multiple_fields)
new_system_stats_df.head(25)
```

```python
query_with_pivot ='''
from(bucket: "System")
  |> range(start:-30d, stop: now())
  |> filter(fn: (r) => r._measurement == "cpu")
  |> filter(fn: (r) => r._field == "usage_user")
  |> filter(fn: (r) => r.cpu == "cpu-total" or r.cpu == "cpu0" or r.cpu == "cpu1")
  |> limit (n:10)
  |> pivot(rowKey:["_time"], columnKey: ["cpu"], valueColumn: "_value")
  |> keep(columns:["_time", "cpu-total", "cpu0", "cpu1"])
'''
```

```python
client = InfluxDBClient(url=urlCloud, token=my_token, org=my_org, debug=False)

final_system_stats_df = client.query_api().query_data_frame(org=my_org, query=query_with_pivot)
final_system_stats_df.head(25)
```

In order to use many anomaly detection, forecasting, and time series analysis libraries, you'll need to convert the "_time" column values to a datetime object. Drop the "result" and "table" columms to clean up the DataFrame further. 

```python
final_system_stats_df["_time"] = pd.to_datetime(final_system_stats_df["_time"].astype(str))
final_system_stats_df = final_system_stats_df.drop(columns=["result", "table"])
final_system_stats_df = final_system_stats_df.set_index("_time")
final_system_stats_df.head()
```

Now our dataframe is ready to be used with other libraries. We can also plot the time seires easily.

```python
final_system_stats_df.plot()
```
